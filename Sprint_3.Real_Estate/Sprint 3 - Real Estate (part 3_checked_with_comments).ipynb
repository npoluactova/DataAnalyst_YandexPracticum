–í—ã–±—Ä–æ—Å–æ–≤ –º–Ω–æ–≥–æ. –ò–∑–±–∞–≤–∏–º—Å—è –æ—Ç –Ω–∏—Ö.

```python
#—Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –ø—Ä–∞–≤–æ–π –∏ –ª–µ–≤–æ–π –≥—Ä–∞–Ω–∏—Ü—ã –¥–ª—è "—É—Å–æ–≤" –Ω–∞ –¥–∏–∞–≥—Ä–∞–º–º–µ —Ä–∞–∑–º–∞—Ö–∞
def iqr_margins(column):
    try:
        e = [max(column.quantile(0.25) - 1.5*(column.quantile(0.75)-column.quantile(0.25)), column.min()),
            min(column.quantile(0.75) + 1.5*(column.quantile(0.75)-column.quantile(0.25)), column.max())]
        return e
    except:
        pass   

#—Å—Ç–æ–ª–±—Ü—ã, –≤ –∫–æ—Ç–æ—Ä—ã—Ö –±—É–¥–µ–º —É–¥–∞–ª—è—Ç—å –≤—ã–±—Ä–æ—Å—ã
columns_to_cleanse = ['total_area', 'kitchen_area', 'living_area', 'last_price', 'floors_total', 'ceiling_height']

#–æ—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ª—å–∫–æ —Ç–µ –∑–∞–ø–∏—Å–∏, –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–ø–∞–¥–∞—é—Ç –≤ –≥—Ä–∞–Ω–∏—Ü—ã "—É—Å–æ–≤" –ø–æ –∫–∞–∂–¥–æ–º—É –∏–∑ —Å—Ç–æ–ª–±—Ü–æ–≤
#–æ—á–∏—â–µ–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ –∑–∞–ø–∏—Å—ã–≤–∞–µ–º –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é final_data
#—Ñ—É–Ω–∫—Ü–∏—é –≤—ã–∑—ã–≤–∞–µ–º –¥–ª—è data, —á—Ç–æ–±—ã –ø–µ—Ä–µ–∑–∞–ø—É—Å–∫ —è—á–µ–π–∫–∏ –≤—Å–µ–≥–¥–∞ –≤ final_data –æ—Å—Ç–∞–≤–ª—è–ª –æ–¥–∏–Ω–∞–∫–æ–≤–æ–µ –∫–æ–ª-–≤–æ —Å—Ç—Ä–æ–∫, 
#—Ç.–∫. –æ—Ç–±—Ä–∞—Å—ã–≤–∞–µ–º –≤—ã–±—Ä–æ—Å—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –∏—Å—Ö–æ–¥–Ω–æ–≥–æ —Å–æ—Å—Ç–æ—è–Ω–∏—è —Ç–∞–±–ª–∏—Ü—ã data
final_data = data
for i in columns_to_cleanse:
    final_data = final_data[
    (final_data[i] >= iqr_margins(data[i])[0]) & 
    (final_data[i] <= iqr_margins(data[i])[1])] 
```
